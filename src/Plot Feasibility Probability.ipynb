```python
idx = [2, 16, 24]
plot_obj_df = df.loc[idx]
plot_obj_df
```

```python
import seaborn as sns
sns.set_style("whitegrid")

fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10), dpi = 200)

sns.lineplot(data=plot_feasible_df, x="rho", y="feasibility (robust)", legend = 'brief', 
            label='Robust Optimization', ax = axes[0]) # marker = "o", markersize = 10
sns.lineplot(data=plot_feasible_df, x="rho", y="feasibility (baseline)", legend = 'brief', 
             label='Baseline', ax = axes[0])
axes[0].set_ylim(0.36,0.45)
axes[0].set_xlabel("", fontsize=15)
axes[0].set_title('Feasible Probability of Robust Modeling with Different rho', fontsize=15)
axes[0].set_ylabel('Feasible Probability', fontsize=15)


sns.lineplot(data=plot_obj_df, x="rho", y="obj_value (robust)", legend = 'brief', 
             label='Robust Optimization', ax = axes[1])
sns.lineplot(data=plot_obj_df, x="rho", y="obj_value (baseline)", legend = 'brief', 
             label='Baseline', ax = axes[1])
axes[1].set_ylim(0,)
axes[1].set_title('Objective Value (Portfolio Standard Deviation) of Robust Modeling', fontsize=15)
axes[1].set_ylabel('Objective Value', fontsize=15)
axes[1].set_xlabel("rho", fontsize=15)

plt.show()
plt.tight_layout()
```
